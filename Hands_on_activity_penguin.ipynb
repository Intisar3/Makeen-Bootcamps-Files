```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Create a plot for mean flipper length with 95% confidence intervals
plt.figure(figsize=(8, 6))
sns.pointplot(data=penguins, x="species", y="flipper_length_mm", ci=95, capsize=0.2)
plt.title("Mean Flipper Length by Species (95% CI)")
plt.xlabel("Species")
plt.ylabel("Mean Flipper Length (mm)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

print(penguins.head())
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.pointplot(data=penguins, x="species", y="flipper_length_mm", ci=95, capsize=0.1, join=False)

plt.title("Mean Flipper Length by Species (95% CI)")
plt.xlabel("Species")
plt.ylabel("Mean Flipper Length (mm)")
plt.show()
```

```python
sns.pointplot(penguins, x="species", y="flipper_length_mm", hue="sex", capsize=0.2, dodge=True, color="orange", join=False)
plt.title("Mean Flipper Lenght by Species")
plt.ylabel("Mean Flipper Length (mm)")
plt.show()
```
