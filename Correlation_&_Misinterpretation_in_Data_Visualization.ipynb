This file contains two mini data analysis projects, both focused on data visualization and correlation:

1. Misleading Correlation Example
Purpose:
To illustrate how correlation between two variables (e.g., ice cream sales and death rate) may exist without causation.

Steps:

Simulates temperature-related data:

Ice Cream Sales increase with temperature

Death Rate also increases with temperature

Other variables: Hydration Level and Accident Rate

Uses a scatter plot to show a misleading correlation between ice cream sales and death rate.

Generates a pair plot to show relationships between all variables for deeper correlation analysis.

Key Insight:
The variables may correlate because they share a common influencing factor (temperature)â€”not because one causes the other.

2. White Coat Perception Study
Purpose:
To analyze which professions are most associated with wearing a white coat.

Steps:

Randomly generates 100 entries of people from four professions: Doctor, Chef, Lab Technician, and Artist.

Simulates the likelihood of them wearing a white coat.

Uses a bar plot to display the percentage of white coat wearers per profession.

Key Insight:
This simple categorical analysis visually communicates perceived associations (e.g., doctors and lab techs wearing white coats more often).

Misleading vs valid correlations

Visualization with matplotlib and seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

np.random.seed(42)

n = 100

temperature = np.random.uniform(20, 40, n)
ice_cream_sales = temperature * 5 + np.random.normal(0, 5, n)
death_rate = temperature * 2 + np.random.normal(0, 10, n)
hydration_level = 100 - temperature + np.random.normal(0, 5, n)
accident_rate = np.random.uniform(10, 30, n) + temperature * 0.3

df = pd.DataFrame({
    "Temperature": temperature,
    "Ice Cream Sales": ice_cream_sales,
    "Death Rate": death_rate,
    "Hydration Level": hydration_level,
    "Accident Rate": accident_rate
})


print(df.head())

plt.figure(figsize=(8, 5))
sns.scatterplot(x="Ice Cream Sales", y="Death Rate", data=df, color="red")
plt.title("Misleading Correlation: Ice Cream Sales vs. Death Rate")
plt.xlabel("Ice Cream Sales")
plt.ylabel("Death Rate")
plt.show()


plt.figure(figsize=(10, 6))
sns.pairplot(df, diag_kind="kde", plot_kws={"alpha": 0.6})
plt.suptitle("Comprehensive Correlation Analysis", y=1.02)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Set random seed for reproducibility
np.random.seed(42)

# Number of data points
n = 100

# Generate dummy data
occupation = np.random.choice(["Doctor", "Chef", "Lab Technician", "Artist"], n, p=[0.25, 0.25, 0.25, 0.25])
wearing_white_coat = np.random.choice([True, False], n, p=[0.6, 0.4])  # Probability of wearing a white coat

# Create DataFrame
df = pd.DataFrame({
    "Occupation": occupation,
    "Wearing White Coat": wearing_white_coat
})

# Calculate the percentage of white coat wearers per occupation
coat_counts = df[df["Wearing White Coat"] == True]["Occupation"].value_counts(normalize=True) * 100

# Bar plot showing the distribution of white coat wearing by occupation
plt.figure(figsize=(8, 5))
sns.barplot(x=coat_counts.index, y=coat_counts.values, palette="viridis")
plt.title("Percentage of Wearing White Coats by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Percentage (%)")
plt.show()
```
