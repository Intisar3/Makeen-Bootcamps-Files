```python
# Step 1: Import the necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load a built-in dataset (e.g., 'tips')
tips = sns.load_dataset('tips')

# Step 3: Calculate the correlation matrix for numerical columns
correlation_matrix = tips.corr()

# Step 4: Set up the heatmap with the specified parameters
plt.figure(figsize=(10, 8))  # Optional: Adjusts the size of the heatmap

sns.heatmap(
    correlation_matrix,          # Using the correlation matrix
    vmin=-1, vmax=1,             # Min and max values of the heatmap
    center=0,                    # Center the heatmap at 0
    cmap=sns.diverging_palette(240, 10, n=500),  # Color palette with 500 shades
    square=True,                 # Use square-shaped cells
    annot=True,                  # Display correlation values in the cells
    fmt=".2f"                    # Format for the correlation numbers
)

# Step 5: Add title and show the plot
plt.title("Correlation Heatmap of Tips Dataset")
plt.show()
```
